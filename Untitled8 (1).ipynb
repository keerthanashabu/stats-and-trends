```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#loading the dataset
file_path = 'Sleep_health_and_lifestyle_dataset (3).csv' 
df= pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
# to find missing values
missing_values= df.isna().sum()
missing_values
```

```python
filling_values={'Sleep Disorder':'no disorder'}
df.fillna(value=filling_values,inplace=True)
```

```python
total_missing_values= df.isna().sum()
total_missing_values
```

```python
# plotting the pie chart
bmi_counts = df['BMI Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%')
plt.title('BMI category',fontsize=12)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Sleep Duration'], df['Quality of Sleep'], alpha=0.5)
plt.xlabel('Sleep duration')
plt.ylabel('Quality of sleep')
plt.title('Sleep duration vs Quality of sleep')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Blood Pressure'], alpha=0.5)
plt.xlabel('Sleep duration')
plt.ylabel('Quality of sleep')
plt.title('Sleep duration vs Quality of sleep')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Stress Level'], df['Physical Activity Level'], alpha=0.5)
plt.xlabel('Sleep duration')
plt.ylabel('Quality of sleep')
plt.title('Sleep duration vs Quality of sleep')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
columns = ['Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level','Heart Rate','Daily Steps']
required_data= df[columns]
correlation_matrix = required_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```
